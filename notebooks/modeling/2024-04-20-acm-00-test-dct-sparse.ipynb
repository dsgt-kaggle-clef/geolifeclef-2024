```python
# import dct
from scipy.fftpack import dct
from matplotlib import pyplot as plt
import numpy as np

# create a sparse signal
for k, cutoff in [(1000, 16), (1000, 8), (1000, 4), (10000, 2)]:
    np.random.seed(0)
    x = np.zeros(k)
    for i in range(np.random.randint(1, 10)):
        x[np.random.randint(0, k)] = 1

    print(x.sum())
    x_dct = dct(x, norm="ortho")
    # plt.plot(x_dct)
    # plt.show()

    # truncate to high frequencies
    x_dct_high = x_dct.copy()
    x_dct_high[:-cutoff] = 0
    x_approx_high = dct(x_dct)
    x_dct_low = x_dct.copy()
    x_dct_low[:cutoff] = 0
    x_approx_low = dct(x_dct)

    plt.plot(x, label="original")
    plt.plot(x_approx_low, label="approx low")
    # plt.plot(x_approx_high, label="approx high")
    plt.legend()
    plt.title(f"DCT approximation of sparse signal k={k}, cutoff={cutoff}, n={x.sum()}")
    plt.show()
```
